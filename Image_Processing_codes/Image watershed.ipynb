```python
import cv2
import numpy as n
from matplotlib import pyplot as plt
```

```python
img=cv2.imread("tiger.jpg")
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
grey=cv2.imread("tiger.jpg",0)
cv2.imshow("grey",grey)
cv2.waitKey(0)
grey2=cv2.cvtColor(grey,cv2.COLOR_BGR2RGB)
plt.imshow(grey2)
```

```python
img_1=cv2.medianBlur(grey,3)
adpt=cv2.adaptiveThreshold(grey,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,101,3)
plt.imshow(adpt)
```
